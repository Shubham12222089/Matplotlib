```python
import matplotlib.pyplot as plt
```

```python
x = ["A","B","C","D","E"]
y=[2,4,3,1,5]
c=["r","y","g","b","black"]
plt.xlabel("Data")
plt.ylabel("Value")
plt.bar(x,y,color=c,width=0.5)
plt.show()
plt.hist(x)
plt.grid()
```

```python
e=[0.0,0.0,0.0,0.5,0.0]
plt.pie(y,labels=x,explode=e,startangle=180,autopct="%0.2f%%")
```

```python
plt.pie(y,labels=x)
plt.pie([1],radius=0.6,colors='w')
```

```python
s = [100,200,300,400,500]
plt.scatter(x,y,sizes=s,color=c)
plt.xlabel("Data")
plt.ylabel("Value")
```
