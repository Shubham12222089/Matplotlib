```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=["c","c++","Java","Python"]
plt.pie(x,labels=y)
```

#explode 

```python
x=[10,20,30,40]
y=["c","c++","Java","Python"]
ex = [0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",startangle=90)
```

```python
#doughnut chart 
```

```python
x=[10,20,30,40]
y=["c","c++","Java","Python"]
plt.pie(x,labels=y,radius=1)
plt.pie([1],radius=0.5,colors="w")
```
