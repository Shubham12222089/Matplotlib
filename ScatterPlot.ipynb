ScatterPlot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,5,6,8,4,5,8]
y = [2,3,5,6,8,9,3,6]
d = [100,200,300,400,500,600,700,800]
colors = ["r", "b", "b", "y", "orange", "purple", "green", "b"]
plt.scatter(x,y,c=colors,sizes=d)
plt.title("Scatter Plot",fontsize = 10)
plt.xlabel("data")
plt.ylabel("value")
plt.show()
```
